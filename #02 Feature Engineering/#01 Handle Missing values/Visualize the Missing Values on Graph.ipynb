```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\INSAKUM46\\AppData\\Local\\Programs\\Python\\Python36\\Scripts\\#04 Machine Learning CookBook\\datasets\\Titanic.csv")
```

```python
values = df.isnull().sum()

nullCount = [x for x in values]

plt.figure(figsize=(16,7))

plt.bar(df.columns,nullCount)
plt.grid()
plt.xlabel("Columns")
plt.ylabel("Rows Count")
```
