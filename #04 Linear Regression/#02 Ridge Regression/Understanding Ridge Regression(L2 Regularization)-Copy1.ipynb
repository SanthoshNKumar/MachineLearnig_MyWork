```python
# https://www.analyticsvidhya.com/blog/2016/01/complete-tutorial-ridge-lasso-regression-python/

import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

X = np.array([i*np.pi/180 for i in range(60,300,4)])
np.random.seed(10)
y = np.sin(X) + np.random.normal(0,0.15,len(X))

data = pd.DataFrame(np.column_stack([X,y]),columns=['X','y'])

plt.scatter(data['X'],data['y'])
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression(normalize=True)
lr.fit(X.reshape(-1,1),y.reshape(-1,1))

y_pred = lr.predict(X.reshape(-1,1))

plt.scatter(X,y)
plt.plot(X,y_pred)
```

```python
for i in range(2,16):  #power of 1 is already there
    colname = 'x_%d'%i      #new var will be x_power
    data[colname] = data['X']**i
print(data.head())
```

```python
def draw(power):
    predictors=['X']
    if power>=2:
        predictors.extend(['x_%d'%i for i in range(2,power+1)])

    linreg = LinearRegression(normalize=True)
    linreg.fit(data[predictors],data['y'])
    y_pred = linreg.predict(data[predictors])

    plt.scatter(X,y)
    plt.plot(data['X'],y_pred)
    
    models_to_plot = [1,3,6,9,12,15]
    
    if power in models_to_plot:
        plt.tight_layout()
        plt.plot(data['X'],y_pred)
        plt.plot(data['X'],data['y'],'.')
        plt.title('Plot for power: %d'%power)
        
        
```

```python
draw(3)
```

```python
draw(5)
```

```python
draw(9)
```

```python
draw(12)
```

```python
draw(12)
```

```python
draw(15)
```

```python
# In the above plots observed that as the power increses model becoming overfitting
```

```python
# Ridge Regression :ridge regression performs L2 regularization, 
# i.e. it adds a factor of sum of squares of coefficients in the optimization objective. 
# Thus, ridge regression optimizes the following:
```

```python
from sklearn.linear_model import Ridge

predictors=['X']
predictors.extend(['x_%d'%i for i in range(2,16)])
                
```

```python

ridgereg = Ridge(alpha=1e-15,normalize=True)

ridgereg.fit(data[predictors],data['y'])

y_pred = ridgereg.predict(data[predictors])

plt.scatter(X,y)
plt.plot(data['X'],y_pred)
```
