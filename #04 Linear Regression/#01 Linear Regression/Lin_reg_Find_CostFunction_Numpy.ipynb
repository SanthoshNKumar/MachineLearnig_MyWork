```python
import numpy as np
import pandas as pd

X = np.matrix(np.array([[1,6.1101],[1,5.5277],[1,8.5186],[1,7.0032]]))
y = np.matrix(np.array([[17.592 ,  9.1302, 13.662 , 11.854 ]]))

theta = np.matrix(np.zeros([1,2]))
alpha = 0.01
iters = 1000
n = len(X)

def costfunc(X,y,theta):
    inner =np.power(((X * theta.T) - y),2)
    return np.sum(inner) / (2*n)

def gradientDescent(X,y,theta,alpha,iters):
    #print("X: \n",X)
    #print("y: \n",y)
    temp = theta
    parameters = theta.shape[1]
    cost = np.zeros(iters)
    
    for i in range(iters):
        error = (X*theta.T) - y
    
        for j in range(parameters):
            term = np.multiply(error,X[:,j])
            temp[0,j] = theta[0,j] - ((alpha / len(X)) * np.sum(term))
    
        theta = temp
        cost[i] = costfunc(X,y,theta)
    
    return theta,cost

gradientDescent(X, y, theta, alpha, iters)
```
