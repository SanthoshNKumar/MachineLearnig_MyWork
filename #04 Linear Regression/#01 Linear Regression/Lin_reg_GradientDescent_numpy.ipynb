```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("C:\\MyWork\\MyLearning\\ML\\Files\\DataSet\\ex1Data1.csv",header=None)

X = data.iloc[:,0] # Read First column
y = data.iloc[:,-1] # Read Second Column
m = len(y) # number of training example


X = X[:,np.newaxis]
y = y[:,np.newaxis]

theta = np.zeros([2,1])
iteration = 10
alpha =0.01
ones = np.ones((m,1))
X = np.hstack((ones,X))

def gradientDescent(X,y,theta,alpha,iterations):
    plt.figure(figsize=(20,10))
    plt.scatter(X[:,1], y)
    for i in range(iterations):
        temp = np.dot(X,theta) - y
        temp = np.dot(X.T,temp)
        theta = theta - (alpha / len(X)) * temp
        #print("GradientDescent for the Iteration :",i)
        #print(theta)
        #print('********')

        plt.plot(X[:,1], np.dot(X, theta))
        plt.show
    #print("*****************************************************")
    #print("Theta:\n",theta)
    #print("*****************************************************")
  
    return theta

theta1 = gradientDescent(X, y, theta, alpha, iteration)
```
