```python
import numpy as np

x = [1,2,4,3,5]
y = [5,7,9,11,13]

theta0 = 0
theta1 = 0

n = len(x)
learningrate = 0.001

ypredicted = list()

for i in range(100):
    ypredicted = np.add(theta0,np.multiply(theta1,x))
    
    costfunction = (1/n)*np.sum(np.square(np.subtract(y,ypredicted)))
    print("CostFunction:",costfunction)
    
    dtheta0 = (-(2/n) * np.sum(np.subtract(y,ypredicted)))
    dtheta1 = (-(2/n) * np.sum(np.multiply(np.subtract(y,ypredicted),x)))
    
    theta0 = theta0 - (learningrate * dtheta0)
    theta1 = theta1 - (learningrate * dtheta1)
```
