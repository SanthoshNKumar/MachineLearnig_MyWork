```python
import pandas as pd

df = pd.read_csv("C:\\MyWork\\MyLearning\\ML\\Files\\DataSet\\BreastCancer.csv")

df.shape # Get the shape of the dataset

df = df.iloc[:,1:-1] # Remove Unused Columns on the dataset

print(df.shape)

# Label Encoding
from sklearn.preprocessing import LabelEncoder

df.iloc[:,0:1] = LabelEncoder().fit_transform(df.iloc[:,0:1])

# Seperate the Inout and output data
X = df.iloc[:,1:]

y = df.iloc[:,0:1]

correlation = df.corr()

# print(Columns in the Correlation)
print(correlation.columns)

correlated_features = list()
```

```python
# Remove the correlated feature withn value greater than 0.9

for i in range(len(correlation.columns)):
    row = "Row :" +str(i)
    colss = 0
    for j in range(i):
        if(abs(correlation.iloc[i,j]) > 0.9):
            colss = "    columns-" + str(j )
            colname = correlation.columns[i]
            print(colname)
            correlated_features.append(colname)
            print(row + "   and" + colss)
```

```python
len(correlated_features)
```

```python
removeFeatures = []

# Select Unique Values from the list
[removeFeatures.append(x) for x in correlated_features if x not in removeFeatures] 

print(removeFeatures)

# Removing Correlated Features 
X = X.drop(removeFeatures, axis =1)

print(X.shape)

X
```
