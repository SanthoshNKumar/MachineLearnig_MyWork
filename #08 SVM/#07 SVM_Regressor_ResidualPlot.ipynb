```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.svm import SVR

df = pd.read_csv("C:\\MyWork\\MyLearning\\ML\\Files\\DataSet\\concrete_data.csv")

data_tr = df
data_tr=data_tr[(data_tr.T != 0).all()]

features = ['cement_component', 'flay_ash', 'water_component','superplasticizer', 'coarse_aggregate']

X_vlas = df[features]

y_vals = df['concrete_strength']

X_train,X_test,y_train,y_test = train_test_split(X_vlas,y_vals,test_size =0.3)
```

```python
model = SVR(kernel='linear')
y_pred = model.fit(X_train, y_train).predict(X_test)

plt.figure(figsize=(10,6))
plt.scatter(range(len(y_test)), y_test, color='black')
plt.plot(y_pred, color='blue', linewidth=3)

print("R2 Score:",r2_score(y_test,y_pred))
```
