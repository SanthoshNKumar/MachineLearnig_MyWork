```python
import numpy as np
from sklearn.preprocessing import StandardScaler,normalize,MinMaxScaler
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris().data

x = iris[:,0:2]
y = load_iris().target
```

```python
plt.scatter(x[:,0][y==0],x[:,1][y==0],c='r')
plt.scatter(x[:,0][y==1],x[:,1][y==1],c='b')
plt.scatter(x[:,0][y==2],x[:,1][y==2],c='g')
```

```python
# Stndarddize the data
x_standadize = StandardScaler().fit_transform(x)
plt.scatter(x_standadize[:,0][y==0],x_standadize[:,1][y==0],c='r')
plt.scatter(x_standadize[:,0][y==1],x_standadize[:,1][y==1],c='b')
plt.scatter(x_standadize[:,0][y==2],x_standadize[:,1][y==2],c='g')
```

```python
x_normalizelL11 = normalize(x, norm='l1',axis=0) 
plt.scatter(x_normalizelL11[:,0][y==0],x_normalizelL11[:,1][y==0],c='r')
plt.scatter(x_normalizelL11[:,0][y==1],x_normalizelL11[:,1][y==1],c='b')
plt.scatter(x_normalizelL11[:,0][y==2],x_normalizelL11[:,1][y==2],c='g')
plt.xlim(0.0025,0.01)
plt.ylim(0.003,0.010)
```

```python
x_normalizelL12 = normalize(x, norm='l2',axis=0) 
plt.scatter(x_normalizelL12[:,0][y==0],x_normalizelL12[:,1][y==0],c='r')
plt.scatter(x_normalizelL12[:,0][y==1],x_normalizelL12[:,1][y==1],c='b')
plt.scatter(x_normalizelL12[:,0][y==2],x_normalizelL12[:,1][y==2],c='g')
```

```python
x_MinMAxScling = MinMaxScaler(feature_range=(0, 1)).fit_transform(x)
plt.scatter(x_MinMAxScling[:,0][y==0],x_MinMAxScling[:,1][y==0],c='r')
plt.scatter(x_MinMAxScling[:,0][y==1],x_MinMAxScling[:,1][y==1],c='b')
plt.scatter(x_MinMAxScling[:,0][y==2],x_MinMAxScling[:,1][y==2],c='g')
```
