```python
import numpy as np
import pandas as pd
outliers=[]

data_1= [10,12,12,13,12,11,14,13,15,10,10,10,100,12,14,13, 12,10, 10,11,12,15,
          12,13,12,11,14,13,15,10,15,12,10,14,13,15,10]
    
threshold=3
mean_1 = np.mean(data_1)
print("Mean:",mean_1)
std_1 =np.std(data_1)
print("Standard Deviation:",std_1)
        
for y in data_1:
    print("*********************")
    print("For the value DataSet:",y)
    z_score= (y - mean_1)/std_1 
    print("z_Score:",z_score)
    if np.abs(z_score) > threshold:
        outliers.append(y)
print(outliers)
```
