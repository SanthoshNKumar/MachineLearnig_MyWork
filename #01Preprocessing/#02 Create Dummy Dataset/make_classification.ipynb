```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
np.random.seed(1000)

X1,y1= make_classification(n_samples=300,n_features=2,n_informative=2,n_redundant=0, n_clusters_per_class=1)
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,5))

for i in range(len(y1)):
    if(y1[i] ==0):
        ax.scatter(X1[i,0],X1[i,1],color='r')
    else:
        ax.scatter(X1[i,0],X1[i,1],color = 'g')
```
