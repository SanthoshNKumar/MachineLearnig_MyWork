```python
# Label Encoder and One Hot Encoder
import numpy as np
import pandas as pd
import scipy
from sklearn import preprocessing 

data = pd.DataFrame(
    {
        "Country" : ['France','Spain','Germany','Spain,','Germany','France','Spain','France,','Germany','France'],
        "Age" : [44,34,25,56,36,47,37,31,49,21],
        "Salary" : [77000,43000,34000,29000,65000,56000,46000,67000,87000,69000],
        "Purchased" :['No','Yes','Yes','No','No','Yes','No','Yes','No','Yes']
    }
)


before = data['Country']

#****************** Implement Label Encoder

# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 

# label_encoder object knows how to understand word labels.
data['Country']= label_encoder.fit_transform(data['Country']) 

#******************

after = data['Country']

pd.DataFrame([before,after])
```
