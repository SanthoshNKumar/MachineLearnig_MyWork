```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("C:\\Users\\INSAKUM46\\AppData\\Local\\Programs\\Python\\Python36\\Scripts\\#04 Machine Learning CookBook\\datasets\\Titanic.csv")

df = df.drop(['Last Name','First Name'],axis=1) # Drop First and Last Column

df.drop(df[df['Embarked'].isnull()].index, inplace = True) # Delete the rows which are null from 'Embarked'

df.head(10)
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(['S','C','Q'])
df['Embarked_encode'] =label_encoder.transform(df['Embarked'])
```
