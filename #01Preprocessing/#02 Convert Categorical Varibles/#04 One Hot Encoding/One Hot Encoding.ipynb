```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv("C:\\Users\\INSAKUM46\\AppData\\Local\\Programs\\Python\\Python36\\Scripts\\#04 Machine Learning CookBook\\datasets\\Titanic.csv")

df = df.drop(['Last Name','First Name'],axis=1) # Drop First and Last Column

df.drop(df[df['Embarked'].isnull()].index, inplace = True) # Delete the rows which are null from 'Embarked'

# Here 'Sex';Embarked; are the categorical columsn
df.head(3)
```

```python
encoder = OneHotEncoder(sparse=False)
X_after = encoder.fit_transform(df[['Sex', 'Embarked']])
```

```python
encoder.categories_
```

```python
df.drop(['Sex','Embarked'],axis=1,inplace=True)

df['female'] = X_after[:,:1]
df['male'] = X_after[:,1:2]
df['Embarked_C'] = X_after[:,2:3]
df['Embarked_Q'] = X_after[:,3:4]
df['Embarked_S'] = X_after[:,4:5]
```
