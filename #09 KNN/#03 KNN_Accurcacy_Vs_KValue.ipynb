```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

wine = load_wine()

X = wine.data
y = wine.target

neighbours = np.linspace(1,10,num=10)
```

```python
accuracy = []

for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(wine.data,wine.target)
    y_pred = knn.predict(wine.data)
    
    accuracy.append(accuracy_score(wine.target,y_pred))
```

```python
plt.plot(accuracy)
plt.xlabel("K Value")
plt.ylabel("Accuracy")
plt.title("KNN")
```
